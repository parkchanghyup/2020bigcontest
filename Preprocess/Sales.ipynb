### 판매실적 데이터 전처리

```python
import pandas as pd
import numpy as np
```

```python
# 실적 데이터 확인
실적 = pd.read_excel('실적데이터.xlsx',header=0,)
print(실적.info())
실적.head()
```

```python
# 시청률 데이터 확인
시청률 = pd.read_excel('시청률데이터.xlsx',header=1,index_col='시간대')
print(시청률.info())
시청률.head()
```

```python
#시청률 데이터 컬럼 처리
시청률.columns = list(pd.date_range(start='20190101', end='20200101').strftime("%Y-%m-%d"))
시청률.head()
```

```python
# 노출(분) 컬럼 널 값 처리
# 노출(분) 컬럼에서 nan 값 인덱스 추출
nan = list(np.where(실적['노출(분)'].isna())[0])


for i in nan:
    실적.iloc[i,1] = int(실적.iloc[i-1,1])
실적.head()
```

```python
# 상품 판매 수량 컬럼 추가
실적['판매수량'] = 실적['취급액']/실적['판매단가']
실적.head()
```

```python
# 판매 단가 범위형 컬럼
def get_str(num):
    if num<50000:
        return '5만원이하'
    elif num < 100000:
        return '10만원이하'
    elif num <300000:
        return '30만원이하'
    elif num <500000:
        return '50만원이하'
    elif num <1000000:
        return '100만원이하'
    else :
        return '100만원이상'
판매단가범위 = [ get_str(x) for x in 실적['판매단가'] ]
실적['판매단가범위'] = 판매단가범위
```

```python
# 광고별 시청률 평균 계산
def get_per(n):
    날짜 = str(실적['방송일시'][n])[0:10]
    시간 = str(실적['방송일시'][n])[11:16]
    num = 0
    for i in range(int(실적['노출(분)'][n])):
        분 = int(시간[-2:])+1
        if 분 == 60 :
            분 = 0
        분  = str(분).zfill(2)
        시간 = 시간[:-2]+분
        num +=시청률.loc[시간,날짜]
    return num/실적['노출(분)'][n]
```

```python
# 시청률 평균 데이터 컬럼으로 추가
시청률평균=[]
for i in range(len(실적)):
    시청률평균.append(get_per(i))
실적['시청률평균']=시청률평균
```

```python
import datetime
def get_time(n):
    몇시 = str(실적['방송일시'][n])[11:13]
    무슨요일  = datetime.date(int(str(실적['방송일시'][n])[:4]),int(str(실적['방송일시'][n])[5:7]),int(str(실적['방송일시'][n])[8:10])).strftime('%A')
    time = [몇시, 무슨요일]
    return time

몇시 =[]
무슨요일 = []
for i in range(len(실적)):
    time = get_time(i)
    몇시.append(time[0])
    무슨요일.append(time[1])

실적['몇시']= 몇시
실적['무슨요일']=무슨요일
실적.head()
```

```python
실적.head()
실적.to_csv('실적.csv',index=False)
```
